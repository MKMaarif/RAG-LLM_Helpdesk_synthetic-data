```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data penggunaan AI Assistant X-3000 per fitur dalam satu minggu
hari = ["Sen", "Sel", "Rab", "Kam", "Jum", "Sab", "Min"]
perintah_suara = [50, 60, 55, 65, 70, 80, 75]
kontrol_smart_home = [30, 35, 32, 40, 38, 45, 50]
pengingat = [20, 25, 22, 27, 30, 35, 32]

# Membuat plot
plt.figure(figsize=(8, 5))
plt.plot(hari, perintah_suara, marker='o', linestyle='-', label="Perintah Suara")
plt.plot(hari, kontrol_smart_home, marker='s', linestyle='--', label="Kontrol Smart Home")
plt.plot(hari, pengingat, marker='^', linestyle='-.', label="Pengingat")

# Menambahkan judul dan label
plt.title("Penggunaan AI Assistant X-3000 Berdasarkan Waktu")
plt.xlabel("Hari")
plt.ylabel("Jumlah Penggunaan")
plt.legend()
plt.grid(True)

# Menampilkan grafik
plt.show()
```

```python
# Data konsumsi daya dan waktu respons dalam berbagai mode operasi AI Assistant X-3000
modes = ["Asisten Personal", "Keamanan", "Hemat Energi", "Navigasi Bebas"]
power_consumption = [50, 70, 30, 90]  # Konsumsi daya dalam Watt
response_time = [200, 250, 300, 180]  # Waktu respons dalam milidetik

fig, ax1 = plt.subplots(figsize=(8, 5))

# Plot konsumsi daya
ax1.set_xlabel("Mode Operasi")
ax1.set_ylabel("Konsumsi Daya (Watt)", color='tab:blue')
ax1.bar(modes, power_consumption, color='tab:blue', alpha=0.6, label="Konsumsi Daya")
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Plot waktu respons
ax2 = ax1.twinx()
ax2.set_ylabel("Waktu Respons (ms)", color='tab:red')
ax2.plot(modes, response_time, marker='o', color='tab:red', linestyle='-', label="Waktu Respons")
ax2.tick_params(axis='y', labelcolor='tab:red')

fig.suptitle("Perbandingan Efisiensi AI dalam Mode yang Berbeda")
fig.tight_layout()

plt.show()
```

```python
# Data penggunaan AI Assistant X-3000 di berbagai lingkungan
lingkungan = ["Rumah", "Kantor", "Industri"]
fitur = ["Asisten Personal", "Keamanan", "Automasi Tugas", "Integrasi IoT"]
data = np.array([
    [80, 60, 75, 50],  # Rumah
    [70, 80, 65, 85],  # Kantor
    [50, 90, 85, 95]   # Industri
])

# Membuat grafik batang horizontal
fig, ax = plt.subplots(figsize=(8, 5))
bar_width = 0.2
x_positions = np.arange(len(fitur))

for i, env in enumerate(lingkungan):
    ax.barh(x_positions + (i * bar_width), data[i], height=bar_width, label=env)

ax.set_yticks(x_positions + bar_width)
ax.set_yticklabels(fitur)
ax.set_xlabel("Persentase Penggunaan (%)")
ax.set_title("Perbandingan Penggunaan AI dalam Rumah, Kantor, dan Industri")
ax.legend()

plt.show()
```
